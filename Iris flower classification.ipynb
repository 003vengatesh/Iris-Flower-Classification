# IRIS FLOWER CLASSIFICATION 

# Importing Libraries 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import r2_score
```

# Loading data 

```python
data = pd.read_csv('IRIS.csv')
```

# Data Preprocessing

```python
encoder = LabelEncoder()
data['new species'] = encoder.fit_transform(data['species'])
```

```python
data = data.drop(columns = 'species', axis = 1)
data.head()
```

# Train Test Split 

```python
x = data.drop(columns = 'new species', axis = 1)
y = data['new species']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

# Model Selection 

## Logistic Regression 

```python
lr_model = LogisticRegression()

lr_model.fit(x_train, y_train)
y_pred = lr_model.predict(x_test)

accuracy_score(y_test, y_pred)
```

## Random Forest Classifier 

```python
rfc_model = RandomForestClassifier()

rfc_model.fit(x_train, y_train)
y_pred = rfc_model.predict(x_test)

accuracy_score(y_test, y_pred)
```

## Support Vector machine 

```python
svc_model = SVC()

svc_model.fit(x_train, y_train)
y_pred = svc_model.predict(x_test)

accuracy_score(y_test, y_pred)
```

## K-Nearest Neighbors 

```python
knn_model = KNeighborsClassifier()

knn_model.fit(x_train, y_train)
y_pred = knn_model.predict(x_test)

accuracy_score(y_test, y_pred)
```

# MODEL 

```python
knn_model = KNeighborsClassifier()

knn_model.fit(x_train, y_train)
y_pred = knn_model.predict(x_test)

print(accuracy_score(y_test, y_pred))
print(r2_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

# Based on the models performance KNeighbors Classifier has been selected as a model for IRIS Data Classification
